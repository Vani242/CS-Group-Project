# Methods

The statistical methods used are Central tendency, Variability, Sample distribution and T-test.

**1. Central Tendency:**

Describes the center or typical value of a dataset. They are used to summarise a sample distribution with a single representative value.

KEY MEASURES:

a. MEAN : The sum of all data points divided by the number of data points

b. MEDIAN: The middle value when the data is sorted.

c. MODE : The most frequently occurring value(s) in the dataset.

**2. Variability:**

Measures how spread out or dispersed the data points are within the sample distribution. High variability indicates more spread, while low variability suggests the data points are closer to the central tendency.

KEY MEASURES INCLUDE:

Range: The difference between the maximum and minimum values.

Variance: The average of the squared differences from the mean.

Standard Deviation (SD): The square root of the variance, indicating the average distance of data points from the mean.

Interquartile Range (IQR): The range of the middle 50% of the data, calculated as Q3−Q1.

**3. Sample Distribution:**

Refers to the distribution of a set of data points collected from a subset of a larger population. It represents how the sample's data points are spread and provides insight into the population it is derived from.

It includes:

Shape: The sample distribution can take various forms, such as normal, skewed, or bimodal.

Relevance: It is used to estimate population parameters, often employing statistical measures like the mean or standard deviation.

**4. T-test:**

T-test can be used to test the difference between the mean of a sample and a hypothesized population mean.

Types of t-test:

One Sample t-test: Compares the mean of a single sample to a known population mean or a hypothesized value.

Independent (two sample) t-test: Compares the means of two independent groups to see if they differ significantly.

Dependent (paired) t-test: Compares the means of two related groups, such as measurements taken before and after an intervention on the same subjects.

Weltch t-test: Welch’s t-test is similar to the independent two-sample t-test, except it does not assume that the two populations that the samples came from have equal variance.

# Sampling distribution

A sampling distribution is created by taking samples from the data, finding their means, and comparing these sample means to the population mean. We also checked the standard error (SE) to see how much the sample means vary from the population mean.

```python
def get_sample_mean(data,column_name):
  sample = data.sample(sample_size)
  sample_mean = data[column_name].mean()
  return round(sample_mean,2)
```

```python
vax_sample_means = []
unvax_sample_means = []
sample_size=50
num_of_sampling=100
for i in range(num_of_sampling):
    vax_sample_mean = get_sample_mean(cov19_df, "Crude vax IR")
    unvax_sample_mean = get_sample_mean(cov19_df, "Crude unvax IR")
    vax_sample_means.append(vax_sample_mean)
    unvax_sample_means.append(unvax_sample_mean)

sample_dict = {"Crude vax IR Sample Means": vax_sample_means, "Crude unvax IR Sample Means": unvax_sample_means}

sampling_df = pd.DataFrame(sample_dict)
sampling_df.head()
```

```python
sam_mean = sampling_df.mean()
SE = sampling_df.std()

vax_sample_mean = sam_mean['Crude vax IR Sample Means']
unvax_sample_mean = sam_mean['Crude unvax IR Sample Means']

print(f"Sample mean vaccinated for Crude vax IR: {round(vax_sample_mean, 2)}")
print(f"Sample mean unvaccinated for Crude unvax IR: {round(unvax_sample_mean, 2)}")

print(f"SE vaccinated for Crude vax IR: {round(SE['Crude vax IR Sample Means'], 2)}")
print(f"SE unvaccinated for Crude unvax IR: {round(SE['Crude unvax IR Sample Means'], 2)}")

pop_mean_vax = cov19_df['Crude vax IR'].mean()
pop_mean_unvax = cov19_df['Crude unvax IR'].mean()

print(f"Population mean vaccinated for Crude vax IR: {round(pop_mean_vax, 2)}")
print(f"Population mean unvaccinated for Crude unvax IR: {round(pop_mean_unvax, 2)}")
```

### Obeservation

The observation shows that the sample mean perfectly matches the population mean, proving the sampling process is accurate and reliable.

The SE is zero because all the sampled values for 'Crude vax IR' and 'Crude unvax IR' are identical, resulting in no variation in the data. Without variability, the standard deviation and the standard error are both zero.

The sampling process showed that the sample means were the same as the population means, and there was no change between the samples (SE = 0). This means the data is consistent. Now, we can move on to hypothesis testing ie t-test to check for differences between the two groups.



# T-test

We used an independent two-sample t-test to check if there is a real difference in infection rates between vaccinated and unvaccinated people in our dataset of 987 entries. This test is ideal for comparing the averages of two separate groups, like "Crude vax IR" (infection rate for vaccinated) and "Crude unvax IR" (infection rate for unvaccinated). It helps us see if the difference in their infection rates is meaningful or just due to chance.

```python
from scipy.stats import ttest_ind
```

```python
group1 = cov19_df["Crude vax IR"]
group2 = cov19_df["Crude unvax IR"]
results_1 = ttest_ind(group1, group2)
print(f"P-Value: {(results_1.pvalue)}")
print(f"T-Statistic:  {round(results_1.statistic,2)}")
```

The p-value is less than 0.05, we can reject the null hypothesis of the t-test and thus, this concludes that there is sufficient difference to say that the the crude vax IR and the crude unvax IR are independent of each other and lead to a different mean value.

To conclude, The t-test gave us a p-value of 3.953869952296504e-22, which shows that we reject the idea that the average infection rates for vaccinated and unvaccinated people are the same.

### Why did we not choose Welch test?




Welch's t-test is used when the variances of the two sample groups are unequal. Hence, as in this case the sample variance is equal for both we didn't opt for Welch test.
